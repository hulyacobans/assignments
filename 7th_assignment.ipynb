```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Ödev 1 -- intihar oranları

# value counts dışında iki kolondan nasıl veri çizdiririm yapamadım? 
# Atıyorum bir kolonda sayılar olsa ve onların toplamını verse gruba göre

df = pd.read_csv('master.csv')

plt.figure(figsize=(15, 8))

plt.subplot(1, 2, 1)
df['generation'].value_counts().plot(kind='bar')
plt.title('Generation Suicide Ratio')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
df['year'].value_counts().plot(kind='bar')
plt.title('Year Suicide Ratio')
plt.ylabel('Count')

plt.show()

```
