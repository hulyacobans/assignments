```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as sts
import statsmodels.api as sm
import warnings


from statsmodels.tsa.stattools import acf
from sklearn.preprocessing import normalize
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.stats.mstats import winsorize
from scipy.stats import jarque_bera
from scipy.stats import normaltest
from scipy.stats import bartlett
from scipy.stats import levene
from sqlalchemy import create_engine
from sklearn import linear_model

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('weatherHis.csv')
df.head()
```

```python
temp = df['Temperature (C)'] - df['Apparent Temperature (C)']
```

```python
X = df[['Humidity','Wind Speed (km/h)']]
X
Y = temp

X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

temp = -2.4381 + 3.0292Humidty + 0.1193WindSpeed

* istatiksel olarak anlamlı özellik değişkenlerimiz. P değerleri sıfır.
* Nem 1 derece arttığında sıcaklık farkı 3.0292 oranında artıyor.
* Her ikisi de artı yönde artırıyor.



```python
hum_wind = df['Humidity'] * df['Wind Speed (km/h)']
df['hum_wind'] = hum_wind

X2 = df[['Humidity', 'Wind Speed (km/h)', 'hum_wind']]
Y2 = temp

X2 = sm.add_constant(X2)

results2 = sm.OLS(Y2,X2).fit()
results2.summary()
```

```python
df = pd.read_csv("trainh.csv")
df.head()
```

```python
df = df.drop([ 'PoolQC','MiscFeature', 'Alley','Fence','FireplaceQu', 'PoolQC', 'MiscFeature'], axis = 1)
```

```python
df.dropna(inplace=True)
```

```python
Y=df["SalePrice"]
X=df[["OverallQual", "TotalBsmtSF", "1stFlrSF", "GrLivArea", 'GarageCars', 'GarageArea']]

regr=linear_model.LinearRegression()
regr.fit(X, Y)
```

```python
X=sm.add_constant(X)
sonuc=sm.OLS(Y, X).fit()
sonuc.summary()
```

```python
print("SalePrice= {:.3f} +\n {:.3f} OverallQual +\n {:.3f} TotalBsmtSF +\n {:.3f} 1stFlrSF +\n {:.3f} GrLivArea +\n {:.3f} GarageCars +\n {:.3f}  GarageArea +\n"
      .format(regr.intercept_, regr.coef_[0], regr.coef_[1], regr.coef_[2], regr.coef_[3], regr.coef_[4], regr.coef_[5]))
```

* 1stFlrSF ve GarageArea istatistiksel olarak anlamlı değiller, p değerleri yüksek
